```python
import numpy as np
import scipy.stats as st
import csv
import matplotlib.pyplot as plt
```

Графическое представление зависимости среднего времени перемножения от размеров матриц

```python
name_file = ["res_500.csv", "res_1000.csv", "res_1500.csv"]
name_time = "matrix_time.csv"
list_number = [500, 1000, 1500]
```

```python
list_time = []
with open(name_time) as file:
    reader = csv.reader(file)
    for row in reader:
        list_time.append(row)
print(list_time)
```

```python
times = []
for i in list_time:
    times.append([int(item) for item in i])
dic_time = {}
dic_time[500] = times[0]
dic_time[1000] = times[1]
dic_time[1500] = times[2]
print(dic_time)
```

```python
list_mean = []
for i in dic_time.keys():
    sample = dic_time[i]
    median_value = np.median(sample)
    mean = np.mean(sample)
    list_mean.append(mean)
```

```python
#создание графика зависимости времени реализации произведения квадратных матриц
plt.plot(list_mean, list_number)
plt.xlabel('Среднее значение времени')
plt.ylabel('Размер матрицы')
plt.show()
```
